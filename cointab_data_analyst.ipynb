importing libraries

```python
import pandas as pd
import numpy as np
```

let's import company dataset first

```python
company_data=pd.read_excel('Company X - Order Report.xlsx')
company_data.rename(columns={'ExternOrderNo':'Order ID'},inplace=True)
company_data
```

```python
company_sku=pd.read_excel('Company X - SKU Master.xlsx')
company_sku
```

```python
company_data.isnull().sum()
```

```python
company_pincode=pd.read_excel('Company X - Pincode Zones.xlsx')
company_pincode
```

```python
company_pincode=company_pincode.rename(columns={'Warehouse Pincode':'Warehouse Pincode as per company','Customer Pincode':'Customer Pincode as per company','Zone':'Zone as per company'})
company_pincode
```

importing courier company data

```python
courier_data=pd.read_excel('Courier Company - Invoice.xlsx')
courier_data.rename(columns={'Zone':'delivery zone as per courier','Billing Amount (Rs.)':'charged bill by courier company','Charged Weight':'total weight as per courier company'},inplace=True)
courier_data
```

```python
courier_data.isnull().sum()
```

```python
output1=pd.concat([company_pincode,courier_data],axis=1,join='inner')
columns=['Order ID','AWB Code','Zone as per company','delivery zone as per courier','total weight as per courier company','Type of Shipment','charged bill by courier company']
output1=output1.reindex(columns=columns)
output1
```

```python
common_id=pd.merge(output1,company_data,on='Order ID')
common_id
```

```python
common_weight=pd.merge(common_id,company_sku,on='SKU')
common_weight
```

```python
common_weight['total weight as per company(kg)']=common_weight['Weight (g)']/1000
```

```python
common_weight
```

```python
weight_slab_as_per_company=[]

for i in common_weight['total weight as per company(kg)']:
    if i<0.500:
        value=0.500
    else:
        value=round(i*2)/2
    weight_slab_as_per_company.append(value)
    
weight_slab_as_per_c_company=[]

for j in common_weight['total weight as per courier company']:
    if j<0.500:
        value1=0.500
    else:
        value1=round(j*2)/2
    weight_slab_as_per_c_company.append(value1)
```

```python
common_weight['weight slab as per company']=weight_slab_as_per_company
common_weight['weight slab as per Courier company']=weight_slab_as_per_c_company
```

```python
common_weight
```

```python
expected_chagre=[]

for i,j,e in zip(common_weight['Zone as per company'],common_weight['Type of Shipment'],common_weight['weight slab as per company']):
    if i=='a' and j=='Forward charges' and e<=0.500:
        values=29.5
    elif i=='a' and j=='Forward and RTO charges' and e<=0.500:
        values=29.5+13.6
    elif i=='a' and j=='Forward charges' and e>0.500:
        values=29.5+23.6
    elif i=='a' and j=='Forward and RTO charges' and e>0.500:
        values=29.5+23.6+23.6
    elif i=='b' and j=='Forward charges' and e<=0.500:
        values=33
    elif i=='b' and j=='Forward and RTO charges' and e<=0.500:
        values=33+20.5
    elif i=='b' and j=='Forward charges' and e>0.500:
        values=33+28.5
    elif i=='b' and j=='Forward and RTO charges' and e>0.500:
        values=33+28.3+28.3
    elif i=='c' and j=='Forward charges' and e<=0.500:
        values=40.1
    elif i=='c' and j=='Forward and RTO charges' and e<=0.500:
        values=40.1+31.9
    elif i=='c' and j=='Forward charges' and e>0.500:
        values=38.9+40.1
    elif i=='c' and j=='Forward and RTO charges' and e>0.500:
        values=40.1+38.9+38.9
    elif i=='d' and j=='Forward charges' and e<=0.500:
        values=45.4
    elif i=='d' and j=='Forward and RTO charges' and e<=0.500:
        values=45.4+31.9
    elif i=='d' and j=='Forward charges' and e>0.500:
        values=45.4+44.8
    elif i=='d' and j=='Forward and RTO charges' and e>0.500:
        values=45.4+44.8+38.9
    elif i=='e' and j=='Forward charges' and e<=0.500:
        values=56.6
    elif i=='e' and j=='Forward and RTO charges' and e<=0.500:
        values=56.6+50.7
    elif i=='e' and j=='Forward charges' and e>0.500:
        values=56.6+55.5
    elif i=='e' and j=='Forward and RTO charges' and e>0.500:
        values=56.6+55.5+55.5
    expected_chagre.append(values)    
```

```python
common_weight['Expected charge as per company']=expected_chagre
```

```python
common_weight['difference between expected charge and billed charge']=common_weight['charged bill by courier company']-common_weight['Expected charge as per company']
```

```python
final_output=common_weight
column_order=['Order ID','AWB Code','total weight as per company(kg)','weight slab as per company','total weight as per courier company','weight slab as per Courier company','Zone as per company','delivery zone as per courier','Expected charge as per company','charged bill by courier company','difference between expected charge and billed charge']
final_output=final_output.reindex(columns=column_order)
final_output
```

```python
final_output.to_excel('output1.xlsx',index=False)
```

# Summary Table 

```python
correct_charge=[]
over_charge=[]
under_charge=[]
under_charge1=[]
for i in final_output['difference between expected charge and billed charge']:
    if i==0:
        correct_charge.append(i)
    elif i>0:
        over_charge.append(i)
    else:
        under_charge.append(i)

for i in under_charge:
    valuess=abs(i)
    under_charge1.append(valuess)
```

```python
summary_table=pd.DataFrame(columns=['Summary','Count','Amount(Rs.)'])
summary=['Total orders where company has been correctly charged','Total Orders where company has been overcharged','Total Orders where X has been undercharged']
count=[len(correct_charge),len(over_charge),len(under_charge)]
amount=[sum(correct_charge),sum(over_charge),sum(under_charge1)]

summary_table['Summary']=summary
summary_table['Count']=count
summary_table['Amount(Rs.)']=amount
```
